```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
diabetes = pd.read_csv("diabetes.csv")
#we are setting the seed value
np.random.seed(42)
#we are taking a sample of 25 observations randomly
random_sample_values = diabetes.sample(n=25)
# we are caluculating the mean of sample and population
mean_random_sample_values = random_sample_values['Glucose'].mean()
mean_population_values  = diabetes['Glucose'].mean()
# we are caluculating the max of sample and population
max_random_sample_values = random_sample_values['Glucose'].max()
max_population_values = diabetes['Glucose'].max()
plt.bar(["Population", "Sample"], [mean_population_values ,mean_random_sample_values ])
plt.title("Mean Values")
plt.ylabel("Glucose")
plt.show()
plt.bar(["Population", "Sample"], [max_population_values,max_random_sample_values])
plt.title("Highest Values")
plt.ylabel("Glucose")
plt.show()
```

```python
import seaborn as sns
values_sample = diabetes['BMI'].sample(n=25)
sample_98_percentile = np.percentile(values_sample, 98)
pop_98_percentile = np.percentile(diabetes['BMI'], 98)
sns.kdeplot(diabetes['BMI'], label='Population')
sns.kdeplot(values_sample, label='Sample')
plt.axvline(sample_98_percentile, color='orange', linestyle='--', label='Sample 98th Percentile')
plt.axvline(pop_98_percentile, color='blue', linestyle='--', label='Population 98th Percentile')
plt.legend()
plt.xlabel('BMI')
plt.ylabel('Density')
plt.show()
print('Population 98th Percentile: ', pop_98_percentile)
print('Sample 98th Percentile: ',sample_98_percentile )
```

```python
blood_pressure_values = diabetes['BloodPressure']
mean_population = np.mean(blood_pressure_values)
std_population = np.std(blood_pressure_values)
percentile_population = np.percentile(blood_pressure_values, 90)
btstrp_mean_value = np.zeros(500)
btstrp_std_value = np.zeros(500)
btstrp_percentile_value = np.zeros(500)
for i in range(500):
    sample = np.random.choice(blood_pressure_values, size=150, replace=True)
    btstrp_mean_value[i] = np.mean(sample)
    btstrp_std_value[i] = np.std(sample)
    btstrp_percentile_value[i] = np.percentile(sample, 90)
fig, axs = plt.subplots(3, 1, figsize=(10, 10))
axs[0].hist(blood_pressure_values, bins=20)
axs[0].axvline(x=mean_population, color='blue', label='Population Mean')
axs[0].axvline(x=np.mean(btstrp_mean_value ), color='orange', label='Bootstrap Mean')
axs[0].set_title('Blood Pressure - Bootstrap Mean')
axs[0].legend()
axs[2].hist(btstrp_std_value , bins=20)
axs[2].axvline(x=std_population, color='blue', label='Population Std')
axs[2].axvline(x=np.mean(btstrp_std_value), color='orange', label='Bootstrap Std')
axs[2].set_title('Blood Pressure - Bootstrap Std')
axs[2].legend()
axs[1].hist(btstrp_percentile_value, bins=20)
axs[1].axvline(x=percentile_population, color='blue', label='Population percentile')
axs[1].axvline(x=np.mean(btstrp_percentile_value), color='orange', label='Bootstrap percentile')
axs[1].set_title('Blood Pressure - Bootstrap Std')
axs[1].legend()
print('Population Mean: %.2f' % mean_population)
print('Bootstrap Mean: %.2f' % np.mean( btstrp_mean_value ))
print('Population Std: %.2f' % std_population)
print('Bootstrap Std: %.2f' % np.mean(btstrp_std_value))
print('Population Percentile: %.2f' % percentile_population)
print('Bootstrap Percentile: %.2f' % np.mean(btstrp_percentile_value))
```
